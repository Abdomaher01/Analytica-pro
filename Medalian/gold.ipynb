```python
import pandas as pd
import numpy as np
df = pd.read_csv("Silver.csv")
```

Dimensional Modeling


```python
dim_users = df[
['follower_count',
'following_count',
'account_age_days',
'verified_flag',
'source_type'
]].drop_duplicates()
dim_users['user_id'] = range(1, len(dim_users)+1)
```

```python
dim_platform = df[['platform']].drop_duplicates()
dim_platform['platform_id'] = range(1,len(dim_platform)+1)
```

```python
dim_crisis = df[['crisis_phase']].drop_duplicates()
dim_crisis['crisis_phase_id'] = range(1,len(dim_crisis)+1)
```

```python
dim_tobic = df[['topic']].drop_duplicates()
dim_tobic ['tobic_id']=range(1,len(dim_tobic)+1)
```

```python
dim_sentiment = df[['sentiment_label']].drop_duplicates()
dim_sentiment ['sentiment_id']=range(1,len(dim_sentiment)+1)
```

```python
dim_emotion = df[['emotion']].drop_duplicates()
dim_emotion ['emotion_id']=range(1,len(dim_emotion)+1)
```

```python
dim_region = df[['region']].drop_duplicates()
dim_region ['region_id']=range(1,len(dim_region)+1)
```

```python
dim_language = df[['language']].drop_duplicates()
dim_language ['language_id']=range(1,len(dim_language)+1)
```

```python
dim_time = df[ [ 'year', 'month', 'day' ] ].drop_duplicates().reset_index(drop=True) 
dim_time['time_id'] = range(1, len(dim_time)+1)
```

Dimensions Merge

```python
df = df.merge(dim_platform,on='platform')
df = df.merge(dim_tobic,on='topic')
df = df.merge(dim_sentiment,on='sentiment_label')
df = df.merge(dim_emotion,on='emotion')
df = df.merge(dim_region,on='region')
df = df.merge(dim_language,on='language')
df = df.merge(dim_crisis,on='crisis_phase')
```

Fact Dimensional

```python
fact_posts = [
    'post_id',
    'platform',
    'crisis_phase',
    'topic',
    'sentiment_label',
    'sentiment_score',
    'emotion',
    'stance_label',
    'claim_type',
    'likes',
    'shares',
    'comments',
    'impressions_estimate',
    'reach_estimate',
    'engagement_velocity',
    'misinformation_probability',
    'credibility_score',
    'uncertainty_index',
    'subjectivity_score',
    'toxicity_score',
    'region',
    'language',
    'is_share',
    'parent_post_id',
    'cascade_depth'
]

fact_posts_final = df[fact_posts].copy()
```

```python
fact_posts_final.to_csv("fact_posts.csv", index=False)
dim_time.to_csv("dim_time.csv", index=False)
dim_platform.to_csv("dim_platform.csv", index=False)
dim_users.to_csv("dim_users.csv", index=False)
dim_sentiment.to_csv("dim_sentiment.csv", index=False)
dim_emotion.to_csv("dim_emotion.csv", index=False)
dim_region.to_csv("dim_region.csv", index=False)
dim_language.to_csv("dim_language.csv", index=False)
dim_crisis.to_csv("dim_crisis.csv", index=False)
```
